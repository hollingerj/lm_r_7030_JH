Load Libraries

```r
library(ggplot2)
```

Importing the data set

```r
data = read.csv('regrex1.csv')
```

Creating the Scatter Plot

```r
ggplot(data , aes(x = x , y = y)) + geom_point()
```

Below is data modeling

```r
model = lm(formula = y ~ x,
           data = data)
sum_stats = summary(model)
sum_stats$r.squared
coef(model)[1]
coef(model)[2]
```

Plotting the linear model

```r
ggplot() + 
geom_line(aes(x = data$x, y = predict(model, newdata = data)), colour = 'green')
```

linear model plus scatter

```r
ggplot() + 
geom_point(aes(x = data$x, y = data$y),
           colour = 'yellow') + 
geom_line(aes(x = data$x, y = predict(model, newdata = data)),
          colour = 'purple')
```
